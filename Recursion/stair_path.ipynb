```python
# ðŸªœ Stair Path Problem in Recursion

# You are standing at the bottom of a staircase with N steps.

# At each move, you can climb:

# 1 step,

# 2 steps, or

# 3 steps

# You want to find how many total ways you can reach the top.
```

```python
# using single sstep and double step 



def stair_path(n):
    # Base cases
    if n == 0:
        return 1
    if n < 0:
        return 0
    
    # Recursive relation
    return stair_path(n-1) + stair_path(n-2) 

print(stair_path(5))  # Example
```

```python

# take 1 step, 2 steps, OR 3 steps at a time.

def stair_path(n):
    # Base cases
    if n == 0:
        return 1
    if n < 0:
        return 0
    
    # Recursive relation
    return stair_path(n-1) + stair_path(n-2)  + stair_path(n-3)

print(stair_path(4))  # Example
```
