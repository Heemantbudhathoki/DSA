```python
def power(a, b):
    if b == 0:           # Base case
        return 1
    return a * power(a, b - 1)  # Recursive call

print(power(2, 5))  # Output: 32
```

```python
def sum_array(arr, n):
    if n==0:
        return 0
    
    else:
        return arr[n-1] + sum_array(arr,n-1)
    
    

print(sum_array([1,2,3,4], 4))
```
