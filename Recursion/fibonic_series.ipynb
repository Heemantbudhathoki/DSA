```python
def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n-1) + fib(n-2)

# Print first 10 Fibonacci numbers
for i in range(10):
    print(fib(i), end=" ")
```

```python
# check array is sorted

def isSorted(arr, n):
    # Base cases
    if n == 0 or n == 1:
        return True
    
    # Check last two elements + recursive call
    return (arr[n-1] >= arr[n-2]) and isSorted(arr, n-1)


# Example usage
arr = [1, 3, 5, 7, 9]
print(isSorted(arr, len(arr)))   # True

```
