```python
def print_maze_paths(maze, x, y, path):
    n = len(maze)
    m = len(maze[0])

    # 1️⃣ Base Case: reached bottom-right ✅
    if x == n-1 and y == m-1:
        print(path)
        return

    # 2️⃣ Check bounds and blocked cells ❌
    if x >= n or y >= m or maze[x][y] == 1:
        return

    # 3️⃣ Recursive call: Move Right ➡️
    print_maze_paths(maze, x, y+1, path + "R")

    # 4️⃣ Recursive call: Move Down ⬇️
    print_maze_paths(maze, x+1, y, path + "D")
```
