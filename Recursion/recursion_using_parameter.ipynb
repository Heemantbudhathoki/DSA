```python
count = 0


def parameter():
    global count
    if count ==3:
        return
    count +=1
    print(15)
    parameter()

parameter()
```

```python
def parameter(x,n):
    if n ==0:
        return 
    print(x)

    parameter(x,n-1)
    
parameter(15,4)
```

```python
# print 1 to n

def print_numbers(n, current=1):
    if current > n:
        return
    print(current)
    print_numbers(n, current + 1)

print_numbers(5)
```

```python
# print n to 1

def print_numbers(n, current=1):
    if current > n:
        return

    print_numbers(n, current + 1)
    
    print(current)
    
    
print_numbers(5)
```
