```python
# ðŸšª Sliding Window Technique (Simple Meaning)

# Sliding Window is a technique used when you need to work with subarrays or substrings that are continuous.

# ðŸ‘‰ Instead of checking every possible subarray again and again,
# ðŸ‘‰ You take one window (a range)
# ðŸ‘‰ Compute something
# ðŸ‘‰ Then slide the window forward by removing the left element and adding the next element on the right.

# This makes the process much faster.
```

```python
# ðŸ’¡ Why use Sliding Window?

# Because it reuses the previous results, so you donâ€™t repeat the same work.

# ðŸªŸ Types of Sliding Windows
# 1ï¸âƒ£ Fixed-size window

# Window size is given (k).
# Example: "Find sum of every subarray of size k".

# 2ï¸âƒ£ Dynamic window

# Window size grows/shrinks based on conditions.
# Example: "Longest substring without repeating characters".
```

```python
# ðŸ§  Example (Fixed Window Size k = 3)

# Array: [1, 3, 5, 2, 8]

# Window moves like:

[1,3,5] â†’ sum = 9
   [3,5,2] â†’ sum = (9 - 1 + 2) = 10
      [5,2,8] â†’ sum = (10 - 3 + 8) = 15
```

```python
# 1st method - brute force
# Example Problem - Maximum Sum of a Subarray with K Elements

def maxSum(arr, n, k):
    max_sum = 0

    for i in range(n - k + 1):
        current_sum = 0
        for j in range(i, i + k):
            current_sum = current_sum + arr[j]

        # this must run AFTER the inner loop finishes
        if current_sum >= max_sum:
            max_sum = current_sum

    return max_sum


# Example
if __name__ == "__main__":
    arr = [5, 2, -1, 7, 3]
    k = 3
    n = len(arr)
    print(maxSum(arr, n, k))   # Output: 6
```

```python
# 2nd methd

def maxSum(arr, k):
    
    # length of the array
    n = len(arr)

    # n must be greater than k
    if n <= k:
        print("Invalid")
        return -1
    
    # Compute sum of first window of size k
    window_sum = sum(arr[:k])
    
    # first sum available
    max_sum = window_sum
    # print(max_sum)
    
    
    
    # Compute the sums of remaining windows by
    # removing first element of previous
    # window and adding last element of
    # the current window.
    for i in range(n-k):
        # print(i)
        window_sum = window_sum - arr[i] + arr[i+k]
        
        if window_sum > max_sum:
            max_sum = window_sum
        else:
            max_sum = max_sum 
        
    return max_sum




if __name__ == "__main__":
    arr = [5, 2, -1, 9, 3,7]
    k = 3
    print(maxSum(arr, k))
```
