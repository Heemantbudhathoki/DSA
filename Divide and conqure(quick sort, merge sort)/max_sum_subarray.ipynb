```python
# using kadans algorithm

def maxSubarraySum(arr):
    
    # Stores the result (maximum sum so far)
    res = arr[0]

    # Maximum sum ending at current index
    maxEnding = arr[0]

    for i in range(1, len(arr)):

        # Replace: maxEnding = max(maxEnding + arr[i], arr[i])
        # Using IF–ELSE
        if maxEnding + arr[i] > arr[i]:
            maxEnding = maxEnding + arr[i]
        else:
            maxEnding = arr[i]

        # Replace: res = max(res, maxEnding)
        # Using IF–ELSE
        if maxEnding > res:
            res = maxEnding
            

    return res


if __name__ == "__main__":
    arr = [2, 3, -8, 7, -1, 2, 3]
    print(maxSubarraySum(arr))
```
