```python
# countdown

def countdown(num):
    if num ==0:
        print("Blast off!")
        
    else:
        print(num)
        countdown(num-1)

num = 10
countdown(num)
```

```python
# factorial 

def factorial(n):
    if n == 1:  # base case
        return 1
    else:
        return n * factorial(n - 1)  # recursive call

print(factorial(4))
```

```python
def mergesort(num):
    # Base case: if list has only one element, it's already sorted
    if len(num) == 1:
        return num

    # Step 1Ô∏è‚É£: Split the list into two halves
    mid = len(num) // 2
    left = num[:mid]
    right = num[mid:]

    # Step 2Ô∏è‚É£: Recursively sort both halves
    left_sorted = mergesort(left)
    right_sorted = mergesort(right)

    # Step 3Ô∏è‚É£: Merge the two sorted halves
    return merge(left_sorted, right_sorted)

 
def merge(left, right):
    result = []   # merged sorted list
    i = j = 0

    # Step 4Ô∏è‚É£: Compare and pick smaller elements
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1 
        else:
            result.append(right[j])
            j += 1

    # Step 5Ô∏è‚É£: Add leftover elements (without using extend)
    while i < len(left):
        result.append(left[i])
        i += 1

    while j < len(right):
        result.append(right[j])
        j += 1

    return result


# üåü Try it!
values = [5, 4, 3, 12, 2, 10, 1]
a = mergesort(values)
print(a)
```

```python
# MergeSort in Python


def mergeSort(array):
    if len(array) > 1:

        #  r is the point where the array is divided into two subarrays
        r = len(array)//2
        L = array[:r]
        M = array[r:]

        # Sort the two halves
        mergeSort(L)
        mergeSort(M)

        i = j = k = 0

        # Until we reach either end of either L or M, pick larger among
        # elements L and M and place them in the correct position at A[p..r]
        while i < len(L) and j < len(M):
            if L[i] < M[j]:
                array[k] = L[i]
                i += 1
            else:
                array[k] = M[j]
                j += 1
            k += 1

        # When we run out of elements in either L or M,
        # pick up the remaining elements and put in A[p..r]
        while i < len(L):
            array[k] = L[i]
            i += 1
            k += 1

        while j < len(M):
            array[k] = M[j]
            j += 1
            k += 1


# Print the array
def printList(array):
    for i in range(len(array)):
        print(array[i], end=" ")
    print()


# Driver program
if __name__ == '__main__':
    array = [6, 5, 12, 10, 9, 1]

    mergeSort(array)

    print("Sorted array is: ")
    printList(array)
```
