```python
# âš¡ Quick Sort â€“ Concept (Bullet Points)
# ðŸ§  Core Idea

# Quick Sort is a Divide and Conquer algorithm.

# Choose one element as a pivot.

# Rearrange (partition) the array so:

# All elements less than pivot go to the left.

# All elements greater than pivot go to the right.

# The pivot goes to its correct sorted position.

# Recursively repeat the process on left + right parts.
```

```python
def partition(num, low, high):
    pivot = num[low]
    i = low
    j = high

    # Main loop
    while i < j:

        # Move i forward until number > pivot
        while num[i] <= pivot and i <= high - 1:
            i += 1

        # Move j backward until number <= pivot
        while num[j] > pivot and j >= low + 1:
            j -= 1

        # Swap when i < j
        if i < j:
            num[i], num[j] = num[j], num[i]

    # Swap pivot with num[j]
    num[low], num[j] = num[j], num[low]

    return j


def quicksort(num, low, high):
    if low < high:
        pi = partition(num, low, high)

        # sort left side
        quicksort(num, low, pi - 1)

        # sort right side
        quicksort(num, pi + 1, high)


# MAIN
num = [4, 1, 7, 6, 3, 2, 8]
quicksort(num, 0, len(num) - 1)
print(num)
```
