```python
def findMax(arr, index):
    # Base case: if it's the last element, return it
    if index == len(arr) - 1:
        return arr[index]
    
    # Recursive case: get max from rest of array
    max_in_rest = findMax(arr, index + 1)
    
    # Return the larger one between current and rest
    if arr[index] > max_in_rest:
        return arr[index]
    else:
        return max_in_rest

# Example usage
arr = [3, 8, 2, 11, 6]
max_value = findMax(arr, 0)
print("Maximum element:", max_value)
```
