```python
def subsetRec(i, arr, res, subset):
    # When index reaches end → store the formed subset
    if i == len(arr):
        res.append(list(subset))
        return

    # INCLUDE current element
    subset.append(arr[i])
    subsetRec(i + 1, arr, res, subset)

    # BACKTRACK → remove last element
    subset.pop()

    # EXCLUDE current element  
    # Skip duplicates to prevent same subset being created again
    while i + 1 < len(arr) and arr[i] == arr[i + 1]:
        i += 1

    subsetRec(i + 1, arr, res, subset)


def subsets(arr):
    arr.sort()                # Sort to handle duplicates
    res = []
    subset = []
    subsetRec(0, arr, res, subset)
    return res


if __name__ == "__main__":
    arr = [1, 2, 2]
    ans = subsets(arr)
    print(ans)
```
