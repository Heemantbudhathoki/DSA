```python
# very imp

# implement lru and lfu chasing using linkedlist

# in exam write only get and put but in company we write all 


class Node:
    def __init__(self,key,value):
        self.key = key
        self.value= value
        self.next = None
        self.prev = None 
        
class LRUCaching:
    def __init__(self, capacity):
        self.dummy_head = Node(0,0) # key and value
        self.dummy_tail= Node(0,0)
        
        self.map = {}
        self.capacity = capacity
        
    
    def put(self,k, v):
        if k in self.map.keys():
            node = self.map.get(k)
            self.remove(node)
            
        
        elif( len(self.map)==self.capacity):
            # remove last recantly used (tail)
            self.remove(self.dummy_tail.prev)
        
        
        newnode = Node(k,v)
        self.insert(newnode)
            
        
    def insert(self, newnode):
        self.map[newnode.key] = newnode

        
        if (self.dummy_head.next == None):
            self.dummy_head.next = newnode
            newnode.prev = self.dummy_head
            newnode.next = self.dummy_tail
            self.dummy_tail.prev = newnode
        
        else:
            newnode.next = self.dummy_head.next
            self.dummy_head.next.prev = newnode
            newnode.prev = self.dummy_head
            self.dummy_head.next = newnode
            
        
        
    def remove(self,node):
        # we also remvoe from node and map 
        self.map.pop(node.key)
        
        node.prev.next = node.next
        node.next.prev = node.prev
        node.next = node.prev = Node
        
    
    def get( self,k):
        node = self.map.get(k)
        
        if (node ==None):
            return -1
        
        self.remove(node)
        self.insert(node)
        
        return node.value
         
         
        
        
cache = LRUCaching(2)
cache.put(1, "A")
cache.put(2, "B")
print(cache.get(1))   # "A" (moves 1 to most recent)
cache.put(3, "C")     # removes key 2 (least recent)
print(cache.get(2))   # -1
print(cache.get(3))   # "C"



        
```
