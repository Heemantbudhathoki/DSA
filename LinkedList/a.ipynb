```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
        
        
class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        
    def append(self,data):
        new_node = Node(data)
        
        if self.head is None:
            self.head = new_node
            self.tail = self.head
        
        else:
            current = self.head
            while current.next is not None:
                current = current.next
                
            current.next = new_node
            self.tail = current.next
            
            
    def insfirst(self,data):
        new_node = Node(data)
        current = self.head
        new_node.next = current
        self.head = new_node
        
        
    def inslast(self,data):
        new_node = Node(data)
        # self.tail.next = new_node
        current = self.head
        while current.next is not None:
            current = current.next
        
        current.next = new_node
        
    
    def ins_spe_post(self,data, pos):
        new_node = Node(data)
        
        if pos == 0:
            current = self.head
            new_node.next =current
            
        else:
            current = self.head 
            count = 0
            
            while current.next is not None and count <pos-1:
                
                current = current.next
                count = count +1
                
            if current is None:
                print("out of range")
                
            else:
                new_node.next = current.next
                current.next = new_node
                
                
            
            
            
            
        
        
            
    def print_out(self):
        
        print("head: ", self.head.data)
        print("tail: ", self.tail.data)
        
        current = self.head
        
        while current is not None:
            print(current.data, end = "->")
            current = current.next
            
        print("None")
        

ll = LinkedList()

ll.append(10)
ll.append(20)
ll.append(30)
ll.append(40)
ll.append(50)
ll.insfirst(60)
ll.inslast(70)
ll.ins_spe_post(100,3)


ll.print_out()
```
