```python
# find the first occurrence of a given elemets x, given that the given array is sorted. if no occurrence of x is found then return -1.



def first_occurrence(arr, x):
    low = 0
    high = len(arr) - 1
    result = -1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == x:
            result = mid        # possible answer
            high = mid - 1      # keep searching on the LEFT side
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1

    return result


# Example
arr = [2,5,5,5,6,6,8,9,9,9]
x = 5
print(first_occurrence(arr, x))   # Output: 1
```

```python
#find the square root of the given non negative value x. round it off to the nearest floor inter value.

# input x = 11
#output = 3
```

```python
def floorSqrt(n):
    
    # start iteration from 1 until the 
    # square of a number exceeds n
    res = 1
    while res * res <= n:
        res += 1
    
    # return the largest integer whose 
    # square is less than or equal to n
    return res - 1

if __name__ == "__main__":
    n = 11
    print(floorSqrt(n))
```

```python

def square_root(x):
    if x < 2:
        return x
    
    low = 1
    high = x//2
    
    ans = 1

    while low <= high:
        mid = (low + high) // 2
        square = mid * mid

        if square == x:
            return mid
        elif square < x:
            ans = mid          # possible floor value
            low = mid + 1
        else:
            high = mid - 1

    return ans


# Example
x = 11
print(square_root(x))   # Output: 3
```
