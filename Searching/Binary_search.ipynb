```python
# Recursion binary search


def binarysearch(arr, target, left=0, right=None):
    if right is None:
        right = len(arr) - 1

    # Base case: if range becomes invalid
    if left > right:
        return -1

    mid = (left + right) // 2

    # Target found
    if arr[mid] == target:
        return mid
    
    # Search in left half
    if target < arr[mid]:
        return binarysearch(arr, target, left, mid - 1)
    
    # Search in right half
    return binarysearch(arr, target, mid + 1, right)


# Example
arr = [-1, 0, 3, 5, 9, 12]
target = 9
print(binarysearch(arr, target))   # Output: 4
```
