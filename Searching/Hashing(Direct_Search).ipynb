```python
# ğŸ” What is Hashing? (Very Simple Meaning)

# Hashing is a technique to store and find data very fast.

# ğŸ‘‰ You give a value (like a name, number, string),
# ğŸ‘‰ a hash function converts it into an index,
# ğŸ‘‰ and the value is stored in that index inside a table.

# Think of it like:

# ğŸ—ï¸ Input (key) â†’ ğŸ›ï¸ Hash Function â†’ ğŸ“¦ Index in Hash Table
```

```python
# ğŸ§  Why Hashing?

# Because it gives O(1) time complexity for:

# âœ”ï¸ Searching
# âœ”ï¸ Inserting
# âœ”ï¸ Deleting

# Much faster than linear search or binary search.
```
