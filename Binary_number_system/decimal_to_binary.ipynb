```python
# decimal to binary


n = 10

result = ""
while n > 0:
    if n % 2 == 0:
        result = "0" + result
    else:
        result = "1" + result
    n = n // 2

print(result)
```

```python
# binary to decimal

binary = "1010"

decimal = 0
power = 0

for digit in reversed(binary):
    decimal += int(digit) * (2 ** power)
    power += 1

print(decimal)
```

```python
# short  easy method 

binary = "1010"
decimal = int(binary, 2)
print(decimal)
```

```python
# Convert -10 to 2's complement (8-bit)

n = 10

# Step 1: Convert +10 to binary manually
binary = ""
temp = n
while temp > 0:
    binary = str(temp % 2) + binary
    temp //= 2

# Make it 8-bit (add leading zeros)
while len(binary) < 8:
    binary = "0" + binary

print("Binary (+10):", binary)

# Step 2: 1's complement (flip bits)
ones = ""
for bit in binary:
    if bit == "0":
        ones += "1"
    else:
        ones += "0"

print("1's complement:", ones)

# Step 3: Add 1 manually (binary addition)
twos = list(ones)
carry = 1

for i in range(7, -1, -1):
    if twos[i] == "1" and carry == 1:
        twos[i] = "0"
        carry = 1
    else:
        twos[i] = str(int(twos[i]) + carry)
        carry = 0
        break

twos = "".join(twos)
print("2's complement (-10):", twos)
```
