```python
def pair_sum(arr,target):
    
    result  = []
    for i in range(len(arr)):
        for j in range(i+1,len(arr)):
            if arr[i] + arr[j] == target:
                result.append(i)
                result.append(j)
                
                return result
            
    return result
    

arr = [2,7,11,15]
target = 9

pair_sum(arr,target)
```

```python
def pair_sum(arr, target):
    
    i = 0
    j = len(arr)-1
    
    result = []
    
    while i<j:
        if arr[i]+arr[j]>target:
            j = j-1
            
        elif arr[i]+arr[j]<target:
            i = i+1
            
        else:
            result.append(arr[i])
            result.append(arr[j])
            return result
        
    return result
            

arr = [2,7,11,15]
target = 26

pair_sum(arr,target)
```

```python
def majority_element(arr):
    n = len(arr)

    for i in range(n):
        freq = 0  # reset count for each element
        
        for j in range(n):
            if arr[i] == arr[j]:
                freq += 1

        if freq > n // 2:      # majority condition
            return arr[i]

    return -1    # no majority



arr = [1, 2, 2, 1, 1]
print(majority_element(arr))

```

[Better Approach - 1] Using Sorting - O(nlog(n)) Time and O(1) Space

```python
def majorityElement(arr):
    
    n = len(arr)
    arr.sort()
    
    # Potential majority element
    candidate = arr[n // 2]

    # Count how many times candidate appears
    count = 0
    for num in arr:
        if num == candidate:
            count += 1
    
    if count > n // 2:
        return candidate
    else:
        return -1


if __name__ == "__main__":
    arr = [1, 1, 2, 1, 3, 5, 1]
    print(majorityElement(arr))
```

[Expected Approach] Using Moore's Voting Algorithm - O(n) Time and O(1) Space

```python
def majorityElement(arr):

    n = len(arr)
    candidate = -1
    count = 0

    # Find a candidate
    for num in arr:
        if count == 0:
            candidate = num
            count = 1
        elif num == candidate:
            count += 1
        else:
            count -= 1

    # Validate the candidate
    count = 0
    for num in arr:
        if num == candidate:
            count += 1

    # If count is greater than n / 2, return 
    # the candidate; otherwise, return -1
    if count > n // 2:
        return candidate
    else:
        return -1

if __name__ == "__main__":
    
    arr = [1, 1, 2, 1, 3, 5, 1]
    print(majorityElement(arr))
```
