write a function to calculate sum and product of alll numbers in an array

```python
def sum_and_product(arr):
    total_sum = 0
    total_product = 1
    
    for num in arr:
        total_sum += num       # add to sum
        total_product *= num   # multiply for product
    
    return total_sum, total_product

# Example usage
arr = [1, 2, 3, 4, 5]
sum_result, product_result = sum_and_product(arr)

print("Sum:", sum_result)
print("Product:", product_result)
```

write a function to swap the max and min nummber of an array.

```python
def swap_max_min(arr):
    if len(arr) == 0:
        return arr  # empty array, nothing to do
    
    # Find indices of max and min
    max_index = 0
    min_index = 0
    for i in range(1, len(arr)):
        if arr[i] > arr[max_index]:
            max_index = i
        if arr[i] < arr[min_index]:
            min_index = i
    
    # Swap max and min
    arr[max_index], arr[min_index] = arr[min_index], arr[max_index]
    
    return arr

# Example usage
arr = [4, 2, 7, 8, 1, 2, 5]
swapped_arr = swap_max_min(arr)
print(swapped_arr)
```

write a function to  print all the unique values in an array.

```python
def print_unique(arr):
    for i in range(len(arr)):
        count = 0
        for j in range(len(arr)):
            if arr[i] == arr[j]:
                count += 1
        if count == 1:
            print(arr[i], end=" ")
    print()

arr = [1, 2, 2, 3, 4, 4, 5]
print_unique(arr)
```
