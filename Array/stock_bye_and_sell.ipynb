stock buy and sell

prices =[7,1,5,3,6,4], return max profit.

leetcode problem 121

```python
def buy_and_sell(price):
    max_profit = 0
    best_buy = float('inf')      # lowest price seen so far

    for p in price:
        # update best_buy
        best_buy = min(best_buy, p)

        # calculate profit if we sell today
        profit = p - best_buy

        # update max profit
        max_profit = max(max_profit, profit)

    return max_profit


price = [7,1,5,3,6,4]
print(buy_and_sell(price))
```
