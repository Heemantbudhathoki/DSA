maximum subarray sum

```python
def maxSubarraySum(arr):
    
    res = arr[0]          # final answer
    maxEnding = arr[0]    # current subarray sum

    for i in range(1, len(arr)):

        # maxEnding = max(maxEnding + arr[i], arr[i])
        if maxEnding + arr[i] > arr[i]:
            maxEnding = maxEnding + arr[i]   # extend previous subarray
        else:
            maxEnding = arr[i]               # start new subarray

        # res = max(res, maxEnding)
        if maxEnding > res:
            res = maxEnding

    return res


arr = [2, 3, -8, 7, -1, 2, 3]
print(maxSubarraySum(arr))
```
