container with monst water 

(brute approach)

height = [1,8,6,2,5,4,8,3,7]

```python
def container(height):
    max_area = 0
    
    for i in range(len(height)):
        for j in range(i + 1, len(height)):
            
            w = j - i                     # width
            h = min(height[i], height[j]) # height of container
            area = w * h                  # area
            
            max_area = max(max_area, area)
    
    return max_area

height = [1,8,6,2,5,4,8,3,7]
print(container(height))
```

container with most water 

optimal approach (2 pointer)

```python
def container_using_two_pointer(height):
    max_area = 0

    i = 0
    j = len(height) - 1

    while i < j:
        w = j - i
        h = min(height[i], height[j])
        area = w * h
        max_area = max(max_area, area)

        # Move the pointer at the smaller height
        if height[i] < height[j]:
            i += 1
        else:
            j -= 1

    return max_area

height = [1,8,6,2,5,4,8,3,7]
print(container_using_two_pointer(height))
```
